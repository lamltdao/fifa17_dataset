```python
# Model for determining best players
# Declaring scores
first = 4
second = 3
third = 2
fourth = 1

# Declaring list of factors of same score
first_list = []
second_list = []
third_list = []
fourth_list = []

# Get score 
def getScore(first_list, second_list, third_list, fourth_list):
    # Declaring variables for sum of factors of each list
    sum_of_first_list = 0 
    sum_of_second_list = 0 
    sum_of_third_list = 0 
    sum_of_fourth_list = 0
    # Total factors
    total_factor = len(first_list) + len(second_list) + len(third_list) + len(fourth_list)
    # Calculate sum of factors of each list
    for i in first_list: 
        sum_of_first_list += first * i
    for i in second_list: 
        sum_of_second_list += second * i  
    for i in third_list: 
        sum_of_third_list += third * i
    for i in fourth_list: 
        sum_of_fourth_list += fourth * i
    # Score = Total Scores / Number of factors
    return (sum_of_first_list + sum_of_second_list + sum_of_third_list + sum_of_fourth_list) / total_factor
# # Plot data
#  '''
#      params {characteristic} String
#      params {preffered_position_list} [String]
#      params {label} String
#      params {num} Number
#  '''
# def plotData(characteristic, preffered_position_list, label, num): 
#     plt.figure(figsize = (15,6))
    
#     for i in preffered_position_list:
        
#     sd_gk = df[].sort_values(characteristic, ascending = False)[:num]
#     x_gk = np.array(list(sd_gk['Name']))
#     y_gk = np.array(list(sd_gk[characteristic]))

#     sb.barplot(x_gk,y_gk, palette = "colorblind")
#     plt.ylabel(lebel)
```

```python
# GK characteristics
first_list = [df.GK_Positioning, df.GK_Diving, df.GK_Reflexes]
second_list = [df.Jumping, df.GK_Handling]
third_list = [df.Composure]
fourth_list = [df.GK_Kicking]
df['gk_Shot_Stopper'] = getScore(first_list, second_list, third_list, fourth_list)
```

```python
# Ranking GK
plt.figure(figsize = (15,6))
sd_gk = df.sort_values('gk_Shot_Stopper', ascending = False)[:5]
x_gk = np.array(list(sd_gk['Name']))
y_gk = np.array(list(sd_gk['gk_Shot_Stopper']))

sb.barplot(x_gk,y_gk, palette = "colorblind")
plt.ylabel("Shot Stopping Score")
```

```python
# Full back characteristics
first_list = [df.Speed, df.Stamina, df.Crossing, df.Acceleration]
second_list = [df.Interceptions, df.Sliding_Tackle]
third_list = [df.Standing_Tackle, df.Marking]
fourth_list = [df.Dribbling, df.Strength]
df['fb_adventurer'] = getScore(first_list, second_list, third_list, fourth_list)
```

```python
# Ranking left back
plt.figure(figsize = (15,6))
sd_fb = df[(df['Club_Position'] == 'LWB') | (df['Club_Position'] == 'LB') | (df['Club_Position'] == 'LM')].sort_values('fb_adventurer', ascending = False)[:5]
x_fb = np.array(list(sd_fb['Name']))
y_fb = np.array(list(sd_fb['fb_adventurer']))

sb.barplot(x_fb,y_fb, palette = "colorblind")
plt.ylabel("Left Back Score")
```

```python
# Ranking right back
plt.figure(figsize = (15,6))
sd_fb = df[(df['Club_Position'] == 'RWB') | (df['Club_Position'] == 'RB') | (df['Club_Position'] == 'RM')].sort_values('fb_adventurer', ascending = False)[:5]
x_fb = np.array(list(sd_fb['Name']))
y_fb = np.array(list(sd_fb['fb_adventurer']))

sb.barplot(x_fb,y_fb, palette = "colorblind")
plt.ylabel("Right Back Score")
```

```python
# CDM characteristics
first_list = [df.Speed, df.Strength, df.Stamina, df.Interceptions]
second_list = [df.Aggression,df.Standing_Tackle, df.Sliding_Tackle]
third_list = [df.Ball_Control, df.Marking, df.Long_Pass]
fourth_list = [df.Heading,df.Short_Pass,df.Balance]
df['cdm_destroyer'] = getScore(first_list, second_list, third_list, fourth_list)
```
